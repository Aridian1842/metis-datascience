```python
import pandas as  pd
import requests
from bs4 import BeautifulSoup
import html5lib
import statsmodels.api as sm
from patsy import dmatrices
import re
```

```python
# Get studio rating data from Metacritic
url = 'http://www.metacritic.com/feature/film-studio-rankings-2014'
headers = {'User-Agent': 'Mozilla/5.0'} # to prevent 403: Forbidden error on above url
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'lxml')
value = soup.find('th', text="Studio (Ranked by Score)")
table = value.parent.parent.parent
rows = table.find_all('tr')[1:]

studio_ranks = {'studio': [],
            'num_films': [],
            'avg_score': [],
            'grade': [],
           }

for row in rows:
    cols = row.find_all('td')
    studio_ranks['studio'].append(cols[1].get_text())
    studio_ranks['num_films'].append(cols[2].get_text())
    studio_ranks['avg_score'].append(cols[7].get_text())
    studio_ranks['grade'].append(cols[8].get_text())
    
studio_ranksData = pd.DataFrame(studio_ranks)
```

```python
studio_ranksData.reindex(columns=['studio', 'num_films', 'avg_score', 'grade'])
```

```python
url = 'http://www.metacritic.com/feature/film-studio-rankings-2014'
headers = {'User-Agent': 'Mozilla/5.0'} # to prevent 403: Forbidden error on above url
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'lxml')
value = soup.find('th', text="2013 Box Office Grosses (in billions)")
table = value.parent.parent.parent
rows = table.find_all('tr')

studio_gross = {'studio': [],
               'domestic_gross': []
               }

for row in rows:
    cols = row.find_all('td')
    studio_gross['studio'].append(cols[1].get_text())
    studio_gross['domestic_gross'].append(cols[2].get_text())

studio_grossData = pd.DataFrame(studio_gross)
```

```python
# Get studio Oscar data from BoxOfficeMojo
url = 'http://www.boxofficemojo.com/oscar/chart/?yr=2013&view=bystudio&p=.htm'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
table = soup.find_all('table')[2]
rows = table.find_all('tr')[1:]

studio_oscars = {'studio': [],
                'domestic_gross': [],
                'noms': [],
                'wins': []
                }

for row in rows:
    cols = row.find_all('td')
    studio_oscars['studio'].append(cols[1].get_text())
    studio_oscars['domestic_gross'].append(cols[2].get_text())
    studio_oscars['noms'].append(cols[5].get_text())
    studio_oscars['wins'].append(cols[6].get_text())
studio_oscarsData = pd.DataFrame(studio_oscars)
```
